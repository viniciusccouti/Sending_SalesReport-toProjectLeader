# Process Automation in a Company System

### Main Goal:

Imagine that for cost control, every day your leader requests a report with all the company's merchandise purchases. Your job as an analyst is to send him an email as soon as you start working, including the total spent, the quantity of products purchased, and the average price of the products.
 
Your leader's email: For this exercise, use your own email as your leader's email.<br>
For the purpose of this exercise imagine that this link: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema is representing your company system

To solve this, we will use pyautogui, a library for automating mouse and keyboard commands.

```python
import pyautogui as py
import time

py.PAUSE = 1 # it gives one second in each command of pyautogui

# step by step

# 1 - entering in the system (with link showed in the description)
py.hotkey('ctrl','t') # open a new chrome tab
py.write('https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema')
py.press('enter')

# loading the browser
time.sleep(5)

# 2 - login 
# the page that is being used is for test purpose only, does not require any specific login or password to get in

# click in login field
py.click(x=861, y=478)
# write login
py.write('viniciuscoutinho')

# going to the password field
py.press('tab') 
py.write('1234')

#access button
py.click(x=918, y=665)
time.sleep(3)
 
# 3 - export database
py.click(x=459, y=562) #clicking in the file
py.click(x=603, y=447) #clicking in the dots
py.click(x=683, y=553) #clicking in download button
time.sleep(3)
```

```python
# 4 - calculate indicators
import pandas as pd

table = pd.read_csv(r'C:\Users\vinicius_coutinho\Downloads\Compras.csv', sep=';') # in order to make this read csv work you need to insert your own path from your computer
display(table)
total_spent = table["ValorFinal"].sum()
quantity = table["Quantidade"].sum()
average_price = total_spent / quantity
print(total_spent)
print(quantity)
print(average_price)

```

```python
import pyperclip
# 5 - send an email to the project leader

# enter in the email
py.hotkey('ctrl','t') # open a new chrome tab to access gmail
py.write('https://gmail.com')
py.press('enter')

time.sleep(3)

# click in write button in order to write a new email

py.click(x=92, y=246)

# fill information in email
#write the recipient
py.write('viniciuscoutinho_@hotmail.com') 
py.press('tab')

py.press('tab') #going to subject field 
pyperclip.copy('Sales Report') #pyperclip was used just to make an example if the words that would be written are in portuguese and have features that are not used in english such as accents
py.hotkey('ctrl','v')

py.press('tab') #writing email body

text = f"""
Dear,

Please check the Sales Report.

Total Spent: ${total_spent:,.2f}
Quantity:{quantity:,}
Average Price:${average_price:,.2f}

If there is any question, please let me know.

Best regards.
"""
pyperclip.copy(text)
py.hotkey('ctrl','v')

# send email
py.hotkey('ctrl','enter')
```
